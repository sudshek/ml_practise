```python
data_train = pd.read_csv('./train.csv')
label = data_train.iloc[0:5000, :1]
images = data_train.iloc[0:5000, 1:]
images
```

```python
train_image, test_image, train_label, test_label = train_test_split(images, label, train_size=0.8, random_state=0)
```

```python
i = 1
img = train_image.iloc[i].as_matrix()
img = img.reshape((28,28))
plt.imshow(img, cmap='gray')
plt.title(train_label.iloc[i,0])
```

```python
plt.hist(train_image.iloc[i])
```

```python
clf = svm.SVC()
clf.fit(train_image, train_label.values.ravel())
clf.score(test_image, test_label)
```
